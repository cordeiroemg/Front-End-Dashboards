# Randomly Mating Populations

We are now ready to consider populations with Mendelian inheritance and begin by inquiring into the frequencies of the different genotypes tha comprise the population. In the last chapter we were interested in the total number of individuals in the population and in different subpopulations (strains). In genetic studies there is usually greater interest in the relative numbers of different genotypes, so it is convenient to express the numbers of different types as proportions of the total. As long as we use deterministic models the total number in the population is not important and we can deal as well with proportions. With stochastic models the number in the population becomes important in determining the extent of random fluctuations and therefore must be taken into consideration.

## 2.1 Gene Frequency and Genotype Frequency

The number of genotypes in a a population greatly exceeds the number of genes and soon becomes enormous. A diploid population with only two alleles at each of 100 loci would have $3^{100}$ possible genotypes. Therefore we effect a great simplification by writting formulae in terms of gene frequencies rather than genotype frequencies. This usually entails some loss of informaiton, for knowledge of the gene frequencies is not sufficient to specify the genotype frequencies; but it is usually possible to do this to satisfactory approximation by introducing other information, such as the mating system and linkage relations.

Furthermore, in a sexually reproducing population the genes are reassorted by the Mendelian shuffle that takes place every generation. The effects of such reassorment are largely transitory, being undone as fast as they are done. At the outse we need formuale to specify the relation between gene frequencies and genotype frequencies. This will also serve to introduce the kind of notation that will be used throughout the book. We start with a single locus with only tw alleles.

Consider a population of $N$ diploid individuals, of which

\begin{align}
&N_{11} \text{ are of genotype } A_1A_1 \\
&2N_{12} \text{ are of genotype } A_1A_2 \\
&N_{22} \text{ are of genotype } A_2A_2 \\
\end{align}


Where $N_{11} + 2N_{12} + N_{22} = N$. It is sometimes convenient to distinguish, among $A_1A_2$ heterozygotes, those that receive $A_1$ from the mother and those that received it from the father. We can do this by designating the two numbers as $N_{12}$ and $N_{21}$. However, in most populations $N_{12} = N_{21}$ and it is not necessary to make any distinctioin between them.

We designate the frequency or **proportion** of the three genotypes by $P_{11}, 2P_{12}, P_{22}$ as follows:

\begin{align}
&A_1A_1 \text{ : } P_{11}= N_{11}/N \\
&A_1A_2 \text{ : } 2P_{12}= 2N_{12}/N \\
&A_2A_2 \text{ : } P_{22}= N_{22}/N \\
\end{align}


From these genotypes frequencies, we can write the frequencies of alleles $A_1$ and $A_2$, which are designated $p_1$ and $p_2$, as follows:

\begin{align}
p_1 = \frac{2N_{11}+ N_{12}}{2N} = P_{11} + \frac{1}{2} \cdot P_{12} &&and&&
p_2 = \frac{2N_{22}+N_{12}}{N} = \frac{1}{2} \cdot P_{12} + P_{22}
\end{align}



Dominance does not always allow us to establish a direct relation between genotypes and phenotypes. Under such circumstances the allele frequency can be measured only of there is some knowledge about the way in which the genes are combined into genotypes in the population. The simplest assumption, and fortunately one that is often very closely approximated in many actual populations, is random mating.



## 2.2 The Hardy-Weinberg Principle


With random mating, the relation between gene frequency and genotype frequency is greatly simplified. By random mating we mean that the mating occur without regard to the genotypes in questioin. In other words, the probability of choosing a particular genotype for a mate is equal to the relative frequency of that genotype in the population.

Notice that it is possible for a population to be at the same time in random mating proportion for some genes and not for others. In many respects, random mating among the different genotypes in the population is equivalent to random combination of the gametes produced by thse individuals. Considering a single locus with two alleles, $A_1$ and $A_2$, which frequencies $p_1$ and $p_2$ and that $p_1 + p_2 =1$.

Hence, the first generation after random mating begins, the proportions of the three genotypes $A_1 A_1, A_1 A_2,$ and $A_2 A_2$ will be $p², 2p_1 p_2$ and $p²$. The principle says that given the gene frequencies and assuming random mating, the zygotic frequencies can be predicted. Also, the Hardy-Weinberg principle is an useful null-hypotheses for frequencies out of the expected equilibrium.

The equilibrium is attained immediately after random mating begins, rather that gradually, is of great importance. This means that it is not necessary to inquire into the past history of a population; if mating was at random during the preceding generation, the principle holds. The only exception is the rather unusual case where the gene frequencies are different in the two sexes.


\begin{align}
\large p_1² + 2p_1 p_2 + p_2² = (p_1 + p_2)^2
\end{align}


The random mating principle can be stated this way: the array of genotype frequencies in the zygotes is given by the square of the array of gene frequencies in the gametes.

Take the example of a rare condition such as the recessive disease, phenylketonuria, which results in mental deficiency is approximately $1$ in $10,000$. What is the proportion of the population are heterozygous carriers of this condition? Let $A$ stand for the normal allele and $a$ for the recessive allele and designate the frequencies as $P_A$ and $P_a$. The Hardy-Weinberg principle shows that $p_a^²$ is 0.001, so $p_a$ is the square root of this or $0.1$. Since $p_A$ and $p_a$ must add up to $1$, $p_A$ is $0.99$. Therefore the frequency of heterozygotes in the population $(2p_A p_a)$ is $2\cdot(0.99)\cdot(0.01) = 0.0198$

This example illustrates the important fact that , when the recessive gene is rare, the number of heterozygous carriers of the gene is enourmously larger than the number of individuals who carry the gene in the homozygous state. In this example the carries outnumber the affected in the ratio of $198$ to $1$. The majory of the sick children would not come from the sick parents but from heterozygous

The Hardy-Weinberg principle if exactly true only in an infinitely large population in which mating is completely at random, but it is approximately correct for the great majority of genes in most cross-fertilizing species.

The principle departures are of two type:

* inbreeding
* assortative mating

An interesting exception to the stated principle is that the equilibrium frequencies are approached gradulally rather than suddenly when the gene is sex-linked and the frequencies are different in the two sexes. Another is a population in which each individual can reproduce both asexually and sexually, as can many plants.

Lets $C$ be the proportion of progeny that are produces clonally or asexually and $S$ the proportion sexually by random mating. Let $P_1$ be the proportion of genotype $AA$ at generation $t$ and $p$ be the frequency of gene $A$. We assume that $C$ is independent of the genotype of the individual. Notice that the genotypes frequency $P_t$ changes from generation to generation wheres the gene frequency $p$ remains constant. In the following generation the $AA$ plants can be produced in two ways: (1) $AA$ plants with frequency $P_t$, reproducing clonally with probability $C$ and yielding a fraction $CP_t$ of the total progeny, and (2) plants that are homogygous or heterozygous for allele $A$ mating at random with probability $S$ and producing $Sp² AA$ progeny. Thus, in the next generation the frequency of $AA$ genotypes will be


\begin{align}
\large P_{t+1} = CP_t + Sp^2
\end{align}

Rearranging and recalling that $C + S = 1$


\begin{align}
&P_{t+1} - p² = CP_t + Sp^2 - p² \\
&P_{t+1} - p² = CP_t - p² (1 -S) \\
&P_{t+1} - p² = C(P_t - p²)
\end{align}


There is the same relation between $P_t$ and $P_{t-1}$ as $P_{t+1}$ and $P_t$


\begin{align}
&P_{t} - p² = C(P_{t-1} - p²) \\
&P_{t} - p² = C²(P_{t-2} - p²) \\
&P_{t} - p² = C^t(P_0 - p²)
\end{align}

Where $P_0$ is the initial frequency of the genotypes $AA$. Unless $C=1$ (exclusively clonal reproduction), the limit of $C^t$ as $t$ increases is $0$, and $P_t$ becomes closer to $p^2$. Each genotype frequency approaches the Hardy-Weinberg value, but only gradully at a rate determined by $C$. The smaller the fraction of asexual progeny, the more rapidly the random-mating proportion are approached.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
p = 0.7  # initial allele frequency for A (hence p^2 is initial Hardy-Weinberg frequency for AA)
P0 = 0.5  # initial genotype frequency for AA (can be different from p^2)
C = 0.8  # clonal reproduction rate
S = 1 - C  # sexual reproduction rate
generations = 50  # number of generations to simulate

# Arrays to store genotype frequencies
P_t = np.zeros(generations)
P_t[0] = P0
p_squared = p ** 2

# Simulate over generations
for t in range(1, generations):
    P_t[t] = C * P_t[t-1] + S * p_squared

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(P_t, label='Genotype frequency of AA ($P_t$)')
plt.axhline(p_squared, color='red', linestyle='--', label='Expected Hardy-Weinberg (p²)')
plt.xlabel('Generation')
plt.ylabel('Frequency of AA Genotype')
plt.title('Convergence to Hardy-Weinberg Equilibrium under Mixed Reproduction')
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```

## 2.4 X-linked Loci

For alelles on the $X$ chromosome the genotype frequencies in the male are the same as the gene frequencies, since the male is haploid for this chromosome. The female frequencies, on the other hand,  are the same as for diploid autosomal loci, since there are two $X$ chromosomes.

This means that there will usually be a large difference in the frequency of an X-linked character in the two sexes. For example, a trait caused by a recessive gene, $a$, with frequency $p$ would have simple have a frequency $p$ in males but $p²$ in females. If the genes were rare, than frequency in males would be be much greater than in females.

A classical example of a human X-linked trait is color blindness. One of the earliest and largest studies was by Waaler (1927). He found that in a group of 18,121 school children in Oslo. about 8% of the boys were color-blind, but only 0.4% of the girls.
We can estimate the expected number by maximum-likelihood method, we let $p$ be the relative frequency of the allele for color blind and $A, B, C,$ and $D$ be the observed numbers of color-blind males (725), normal males (8324), color-blind female (40) and normal females (9032), respectively. Then, if $m$ is the proportion of males and $N$ of the total observed number $N= A+B+C+D = 18,121),the probability of the observed results is


\begin{align}
Prob = K(mp)^A \left[ m(1-m) \right]^B
\left[ (1-m)p²  \right]^C \left[ (1-m)(1-p²)  \right]^D
\end{align}

Where $K$ is a constant determined by $A, B, C,$ and $D$. Letting $L= log Prob$

\begin{align}
L = (A + B) \cdot log \:m + (C + D) \cdot log \: (1-m) + (A + 2C) \cdot log \:p + (B + D) \cdot log \: (1 - p) + D \cdot log( 1 + p)
\end{align}


solving there equations,

\begin{align}
m = \frac{A + B}{N} = \frac{9049}{18121}
\end{align}

Differentiating with respect to $m$ and $p$, and equating to zero, we obtain:

\begin{align}
\frac{\partial L}{\partial m} = \frac{A + B}{m} - \frac{C + D}{1 - m} = 0
\end{align}

and

\begin{align}
\frac{\partial L}{\partial p} = \frac{A + 2C}{p} - \frac{B + D}{1 - p} + \frac{D}{1 + p} = 0
\end{align}

Solving these equations, we have

\begin{align}
m = \frac{A + B}{N} = \frac{9049}{18121}
\end{align}



As expected, and

\begin{align}
p= \frac{ -B + \sqrt{B² + 4(A + 2C) \left[ A + B + 2(C + D)\right]}}{2 \left[ A + B + 2(C + D) \right]}
\end{align}


The expected numbers in the four classes are $Nmp$, $Nm(1 - p)$, $N(1 - m)p^2$ and $N(1 - m)(1 - p^2)$

```python
import math

# Observed values
A = 725   # Color-blind males
B = 8324  # Normal males
C = 40    # Color-blind females
D = 9032  # Normal females

# Total population
N = A + B + C + D

# Compute allele frequency p using the closed-form formula
numerator = -B + math.sqrt(B**2 + 4 * (A + 2 * C) * (A + B + 2 * (C + D)))
denominator = 2 * (A + B + 2 * (C + D))
p = numerator / denominator

print(f"Estimated allele frequency p for the X-linked trait: {p:.4f}")
```

## 2.5 Different Initial Gene Frequencies in the Two Sexes


Usually the gene frequencies are the same in the two sexes, but under some unusual circumstances they may be different. One possibility is in animal or plant breeding where the original hybrids are made up of males of one strain and females of another. Another is in a natural population where most of the migrants are of one sex.

For an autosomal locus, the Hardy-Weinberg frequencies are attained, but after a delay of one generation. Consider a population in which the frequencies are the alleles $A_1$ and $A_2$ are $p_1^*$ and $p_2^{**}$ in females. If the gametes are combined at random, then in the next generation the frequency of gene $A_1$ in both sexes is

\begin{align}
&p_1 = p_1^* p_1^{**} + \frac{1}{2} p_1^* p_2^{**} + \frac{1}{2}p_1^{**} p_2^{*} \\
& p_1 = \frac{1}{2}(p_1^* + p_1^{**})
\end{align}

And in all following generations the genotypes are in the proportion $p_1^2$, $2p_1 p_2$ and $p_2^2$. As expected, the gene frequency is the unweighted average of what it was originally in the two sexes.

With an X-linked locus, starting out with different allele frequencies in the two sexes, the situation is quite different. The equilibrium, instead of being attained in two generations as for an autosomal locus, is reached only gradually.

Consider a particular X-linked allele, $A$, in a multiple-allelic series. Let the frequency of this allele in generation $t$ be $p_t^*$ in males and $p_t^{**}$ in females. Since a male always get his $X$ chromosome from his mother, (1) the allele frequency in males will alwasys be what it was in females a generation earlier. Likewise, (2) the frequency in females will be the average of the two sexes in the preciding generation, since each sex contributes one $X$ chromosome. Finally, (3) the mean frequency of the gene will be the weighted average of that in the two sexes, attaching twice as much weight to the female frequency quantity must be a constant, since the mean frequency of the gene does not change. These statements may be stated mathematically as follows

Male frequency = female frequency from previous generation
(1) $p_t^* = p_{t-1}^{**}$

Female frequency = average of both sexes from previous generation
(2) $p_t^{**} = \frac{1}{2} p_{t-1}^{*} + \frac{1}{2} p_{t-1}^{**}$

Mean population frequency stays constant
(3) $\bar p_t = \frac{1}{3} p_{t-1}^{*} + \frac{2}{3} p_{t}^{**} = \frac{1}{3} p_{t-1}^{*} + \frac{2}{3} p_{t-1}^{**} = \bar p_{t-1} = \bar p$


Where $\bar p$ is a constant throughout the process. From the third equation we have $p_{t-1}^* = 3 \bar p - 2 p_{t-1}^{**}$. Substituting this into the second equation, after some algebraic rearangement we obtain a relation between the gene frequency in the female sex from generation to generation.

\begin{align}
p_t^{**} - \bar p = - \frac{1}{2}(p_{t-1}^{**} - \bar p)
\end{align}

Since the same relationship holds for $p_{t-1}^{**}$ and $p_{t-2}^{**}$,

\begin{align}
p_t^{**} - \bar p = \left[ - \frac{1}{2} \right]^2 (p_{t-2}^{**} - \bar p) \\
p_t^{**} - \bar p = \left[ - \frac{1}{2} \right]^3 (p_{t-3}^{**} - \bar p) \\
p_t^{**} - \bar p = \left[ - \frac{1}{2} \right]^t (p_{0}^{**} - \bar p) \\
\end{align}


Where $p_0^{**}$ is the allele frequency in females in the initial generation. Notice that these formulae do not depend on equal numbers of males and females in the population.

Thus the approach to equilibrium is gradual. Since $ (-1/2)^t$ (alternating convergence) approaches to $0$ as $t$ becomes larger, the gene frequency must approach $\bar p$ as a limit. In each successive generation the gene frequency in females is only half as far from the final value, $bar p$, as it was in the previous generation, but in the opposite direction. The population moves toward equilibrium in a zig-zag manner, like a series of damped vibrations.



```python
import matplotlib.pyplot as plt

# Initial allele frequencies
p0_male = 0.1     # initial frequency in males
p0_female = 0.9   # initial frequency in females

# Calculate the constant average frequency
p_bar = (1/3) * p0_male + (2/3) * p0_female

# Number of generations to simulate
generations = 20

# Lists to store frequencies over generations
male_freqs = [p0_male]
female_freqs = [p0_female]
mean_freqs = [p_bar]

# Simulate generations
for t in range(1, generations + 1):
    # Male frequency in generation t is female frequency from t-1
    p_male = female_freqs[-1]
    # Female frequency is average of previous male and female frequencies
    p_female = 0.5 * (male_freqs[-1] + female_freqs[-1])

    male_freqs.append(p_male)
    female_freqs.append(p_female)
    mean_freqs.append(p_bar)  # stays constant

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(male_freqs, label="Male Allele Frequency ($p^*_t$)", marker='o')
plt.plot(female_freqs, label="Female Allele Frequency ($p^{**}_t$)", marker='s')
plt.axhline(y=p_bar, color='gray', linestyle='--', label="Mean Allele Frequency ($\\bar{p}$)")
plt.title("Convergence of X-linked Allele Frequencies Over Generations")
plt.xlabel("Generation")
plt.ylabel("Allele Frequency")
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```

## 2.6 Two Loci

Previously we noted that Hardy-Weinberg relation was attained in a single generation, irrespective of the number of alleles. But it is possible for each of two loci to be in random-mating frequencies, yet for them not to be in equilibrium with each other. We shall see that, contrary to the situation with a single locus, the equilibrium relation between two or more loci is not attained immediately, but over a number of generations. We still assume that there is no selection.

Consider two linked loci, with the recombination frequency between them equal to $c$. Unlinked loci can be regarded as a special case where $c$ is $1/2$ Assume that there are $n$ alleles at the locus $A$ and $m$ at the locus $B$.


\begin{array}{c} \hline
A \text{ alleles} & A_1 & A_2 & A_i \\
\text{frequency} & p_1 & p_2 & p_i \\
B \text{ alleles} & B_1 & B_2 & B_k \\
\text{frequency} & q_1 & q_2 & q_k \\ \hline
\end{array}


Let $P_t(A_i B_k)$ be the frequency of the chromosome $A_i B_k$ among the gametes produced in generation $t$. We want to inquire into the frequency of this chromosome in successive generations. For brevity, we shall designate $P_t(A_i B_k)$ simply as $P_t$.

The key idea here is that if two loci are not in equilibrium, then their **haplotype frequencies** are not equal to the product of their allele frequencies.

\begin{align}
P_t(A_i B_k) \ne p_i \cdot q_k
\end{align}

And the deviation from equilibrium is measured as

\begin{align}
D_t = P_t - p_i q_k
\end{align}


The gametes produced in generation $t$ may be of two kinds; they may be the product of a recombination, with probability $c$, or they may not have been involved in a recombination at the preceding meiosis, with probability $( 1- c)$. If no recombination has occured, the probability of the $A_i B_k$ chomosome is the same as it was in the preceding generation, which will be desinated as $P_{t-1}$. If a crossover has occurred, the $A$ and $B$ alleles in a gamete produced by the $t$th generation will have come from different gametes in the $t - 1$th generation; that is, one from the egg and one from the sperm that produced the individual whose gamete we are discussing. Since we are assuming random mating, the egg and sperm are independent, and therefore the $A$ locus and the $B$ locus are independent. Thus, in the gamete produced the the $t$th generation, the probability of the $A$ being $A_i$ and $B$ allele being $B_k$ is simply the product of the two gene frequencies, $p_i q_k$.


\begin{align}
P_t = (1 - c)P_{t-1} + c p_i q_k
\end{align}

After substituting $p_i q_k$ from both sides

\begin{align}
P_t - p_i q_k = (1 - c) \cdot (P_{t-1} - p_i q_k) \\
P_t - p_i q_k = (1 - c)^2 \cdot (P_{t-2} - p_i q_k)\\
P_t - p_i q_k = (1 - c)^3 \cdot (P_{t-3} - p_i q_k) \\
P_t - p_i q_k = (1 - c)^t \cdot (P_{0} - p_i q_k)
\end{align}


Where $P_0$ is the initial value of the frequency of chromosome $A_i B_k$.

This is an exponential decay model of the linkage disequilibrium over generations

\begin{align}
D_t = ( 1- c)^t D_0
\end{align}

Thus the frequency of the $A_i B_k$ chromosome gets closer to the value $p_i q_k$; each generation the departure from the final value is reduced by a fraction equal to the recombination value. For unlinked loci, the chromosome frequency goes half way to the equilibrium value each generation.



```python
import matplotlib.pyplot as plt
import numpy as np

# Initial LD: D0 = P0 - p_i * q_k
D0 = 0.1  # Initial departure from equilibrium

# Recombination rates
c_values = [0.5, 0.2, 0.1]
generations = np.arange(0, 9)

# Compute D_t for each recombination rate
ld_trajectories = {}
for c in c_values:
    D_t = D0 * (1 - c) ** generations
    ld_trajectories[c] = D_t

# Plotting
plt.figure(figsize=(10, 6))
for c, D_t in ld_trajectories.items():
    plt.plot(generations, D_t, marker='o', label=f'c = {c}')

plt.title('Decay of Linkage Disequilibrium Over Generations')
plt.xlabel('Generation (t)')
plt.ylabel(r'$P_t - p_i q_k$ (Linkage Disequilibrium $D_t$)')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.7)
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```

The equilibrium value is approached only gradually, and theoretically is never attained. For comparison it is convenient to speak of the time required to go halfway to the final value, in the same way that one speaks of the halflife or median life of a radioactive element.

The median equilibrium time is given by solving for $t$ the equation

\begin{align}
(1 - c)^t = \frac{1}{2}
\end{align}

Which leads to

\begin{align}
t = \frac{log \: \frac{1}{2}}{log ( 1- c)}
\end{align}

When $c$ is small, $log_e (1 -c)$ is very close to $-c$. So, using logs to the base $e$, $log (1/2) = - 0.693$ and $t = 0.693/c$ approximately for small values of $c$. For example, with independent gene loci $c = 1/2$, the median time is 1 generation. When $c = 0.1$, the approximation is satisfatory and $t$ is $0.693/0.1$, or about 7 generations. When, $c = 0.01$, the time is about 69 generations and when $c = 0.001$, about 693 generations.

We can summarize the main results of this section in two statements:
* The approach to equilibrium between coupling and repulsion phases of the gametes (usually called gametic phase equilibrium or 'linkage' equilibrium) is gradual. The rate of approach depends on the rate of recombination between the two loci. For independent loci is 50% per generation
* We can broaden the meaning of the Hardy-Weinberg principle by regading chromosomes or ganmes rather than single genes. With random mating the frequency of any diploid genotype is given by the appropriate term in the expansion of the square of the array of gamete frequencies. These frequencies are attained immediately if the game frequencies are the same in both sexes.


\begin{array}{c} \hline
Genotype & Gamete freq & Equilibrium \\ \hline
A_i B_k| A_i B_k & P_{ik}^2 & p_i^2 p_k^2  \\
A_i B_k| A_j B_l & 2P_{ik}P_{jl}  & 2p_i p_j q_k q_l \\
A_i B_l| A_j B_k & 2P_{il}P_{jk}   & 2p_i p_j q_k q_l \\ \hline
\end{array}


Crossing over in the two sexes is not necessarily the same. The formula given earlier can earily be modified to take this into consideration where $c_m$ and $c_f$ stand for the recombination fraction in males and females.

\begin{align}
&P_t = \frac{1}{2} \left[ (1 - c_m)P_{t-1} + c_m p_iq_k \right] + \frac{1}{2} \left[ (1 - c_f) P_{t-1} + c_f p_i q_k \right] \\
&P_t = (1 - \bar c) P_{t-1} + \bar c p_i q_k
\end{align}

Where $\bar c = \frac{1}{2} (c_m + c_f)$


Therefore, when crossing over differs in males and females it is sufficient to replace $c$ in equations by mean value in the two sexes. In Drosophila and Bombyx, where meiotic crossing over is restricted to one sex, the approach to equilibrium is just half as fast as it would be if it occurred at the homogametic rate in both sexes.

A population that has been mating at random for many generations, a strong association between traits is not due to linkage unless the amount of recombination between the responsible loci is very small.Such persisting association as are found in natural populations are probably due to pleiotropy or to recent amalgamation of population with incomplete mixing, that is that say nonrandom mating.

```python
# Re-import necessary libraries after code execution environment reset
import numpy as np
import matplotlib.pyplot as plt

# Define recombination values
c_values = np.array([0.001, 0.005, 0.01, 0.02, 0.05, 0.1, 0.2, 0.5])

# Calculate median time to equilibrium: t = log(1/2) / log(1 - c)
t_median = np.log(0.5) / np.log(1 - c_values)

# Approximation for small c: t ≈ 0.693 / c
t_approx = 0.693 / c_values

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(c_values, t_median, label='Exact Median Time', marker='o')
plt.plot(c_values, t_approx, label='Approximation ($0.693 / c$)', linestyle='--', color='gray')
plt.xlabel('Recombination Rate (c)')
plt.ylabel('Median Time to Equilibrium (Generations)')
plt.title('Median Time to Reach Linkage Equilibrium vs Recombination Rate')
plt.legend()
plt.grid(False)
plt.xscale('log')
plt.yscale('log')
plt.tight_layout()
plt.show()
```

## 2.9 Subdivision of a Population: Wahlund's Principle


In nature or in domestic animals and plants the populations is often structured. One possibility is that is is divided into subpopulations or isolates, between which there is partial or complete isolation. In such strains the gene frequencies may diverge, either because of different environments that favor different genotypes, or simply by chance if the subpopulations are small. We inquire into the effect of amalgamation of previously isolated subpopulations.

The key relationship is expressed in terms of the variance in gene grequencies among the subpopulations. The formula was discussed by Wahlund (1928) and is often called Wahlund's principle.

Imagine that there are $k$ subpopulations, completely isolated from each other and of size $n_1, n_2, ..., n_k$. Mating within each subpopulation is assumed to be at random. Let the frequency of allele $A$ be $p_1, p_2, ..., p_k$ in these subpopulations. Then the mean proportion of $AA$ homozygotes in the whole population is

\begin{align}
mean(p^2) = \frac{n_1p_1^2 + n_2p_2^2 + ... + n_kp_k^2 }{n_1 + n_2 + ... + n_k}
\end{align}


Nos suppose that these populations are pooled into single panmictic unit. The average frequency of the $A$ allele is now (as before) $\bar p$, weighted average of the frequencies in the different populations. Then the proportion of $AA$ homozygotes in the pooled population after one generation of random mating is $\bar p^2$. The mean allele frequency across the populations is

\begin{align}
\bar p = \frac{n_1p_1 + n_2p_2 + ... + n_kp_k }{n_1 + n_2 + ... + n_k}
\end{align}

Recall that the variance is $V_p =  mean(p^2) -  \bar p^2 $. Hence

\begin{align}
\bar p^2 = mean( p^2) - V_p
\end{align}

Where $V_p$ is the variance in the frequency of the gene $A$ among the $k$ sub-populations.

This explains why the proportion of individuals with recessive traits is reduced by migration between previously isolated populations. Since the variance is always positive, there will always be a decrease unless the gene frequency is identical in the subpopulations. The magnitude of the decrease will depend on the diversity of frequencies among the populaitons, as measured by the variance.

The previous discussion has referred to a sintuation where two or more populations are pooled, then mating occur without regard to the origin of the individuals. The situation is somewhat different if the first matings are all between individuals from different populations. We shall consider two populations.

If $p_1$ and $p_2$ represent the frequency of the $A$ gene in the two populations, the proportion of $AA$ homozygotes in the $F_1$ hybrids is $p_1 q_1$. The gene frequency in the $F_1$ is the mean of the two parent population frequencies, or $\bar p = (p_1 + p_2)/2$. The variance in the two original population (equally weighted) is

\begin{align}
V_p = \frac{1}{2}p_1^2 + \frac{1}{2}p_2^2 - \bar p^2 = \frac{1}{4} (p_1 - p_2)^2
\end{align}


Note that $\bar p^2 - V_p = p_1 p_2$, which is the frequency of $AA$ homozygotes in the $F_1$ population.


For comparison, the proportion of $AA$ homozygotes in the three populations is:

1. Separate populations: $ \bar p^2 + V_p$
2. $F_1$ population: $ \bar p^2 - V_p$
3. $F_2$ and later: $ \bar p^2$


Hybridization between two populations causes an initial decrease in homozygosity, followed by a rise to a point halfway between. This argument does not consider linkage, the effect of which is to slow the approach to the final value.

```python
import matplotlib.pyplot as plt
import numpy as np

# Example allele frequencies
p1 = 0.2
p2 = 0.8

# Mean allele frequency
p_bar = (p1 + p2) / 2

# Variance in allele frequencies
Vp = 0.25 * (p1 - p2)**2

# Homozygosity levels
homo_structured = p_bar**2 + Vp        # before mixing
homo_F1 = p1 * p2                      # F1 generation (crosses)
homo_F2 = p_bar**2                     # F2 and later generations

# Labels and values
labels = ['Structured (before mixing)', 'F1 Hybrids', 'F2 and Later']
values = [homo_structured, homo_F1, homo_F2]

# Plotting
plt.figure(figsize=(8, 5))
bars = plt.bar(labels, values, color=['steelblue', 'seagreen', 'darkorange'])
plt.title('Proportion of AA Homozygotes Across Population Types')
plt.ylabel('Homozygosity (AA Frequency)')
plt.ylim(0, 1)

# Annotate values
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.02, f'{yval:.3f}', ha='center', va='bottom')

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## 2.10 Random-mating Proportions in a Finite Population

The Hardy-Weinberg proportion are realized exactly only in a infinite population. For one thing, a finite populaiton is subject to chance deviations from the expected proportions. There is also a systematic bias because of the discreteness of the possible numbers of different genotypes. THe bias can become important if there are a number of individually very rare alleles. Consider a population of size $N$.

Since we are considering diploid populations, there are $2N$ genes per locus. Let $p_i$ be the proportion of allele $A_i$ in the population; hence there are $2N \cdot p_i$ representatived of the $A_i$ allele. Then we regard the gygotes as made up by combining these $2N$ alales at random pairs. The probability of drawing an $A_i$ allele is $ (2N \cdot p_i )/ 2N$; after this is done, the probability of drawing another $A_i$ allele from the remaining genes is $(2N \cdot p_i -1)/(2N -1)$. Thus the expected proportion of $A_iA_i$ individuals, given that there are exactly $2N \cdot p_i A_i$ alleles is


\begin{align}
P(A_i, A_i) = \frac{2Np_i}{2N} .\frac{2Np_i -1}{2N-1} = p_i^2 - p_i (1 - p_i).f
\end{align}

Where $f = 1/(2N -1)$. Likewise the expected proportion of $A_iA_j$ heterozygotes is

\begin{align}
P(A_i, A_i) = \frac{2Np_i}{2N} .\frac{2Np_j}{2N-1} +  \frac{2Np_j}{2N} .\frac{2Np_i}{2N-1}= 2p_i p_j(1+f)
\end{align}


Thus the heterozygotes are increased by a fraction $f = 1/(2N -1)$ and the homozygotes are correspondingly decreased, in comparison with the proportions in an infinite populaiton with the same allele frequencies.

Heterozygosity is slightly elevated and homozygosity slightly reduced in a finite population - in contranst to what drit does over time (which reduces heterozygosity). This effect is purely due to sampling bias in genotype formation. The bias matters most for rare alleles. If $p_i$ is small, the correction becomes relative larger. For small $N$, the correction becomes biologically significant.

```python
import matplotlib.pyplot as plt
import numpy as np

# Allele frequency
p = 0.05
q = 1 - p

# Range of population sizes
N_values = np.arange(10, 1001, 10)
f_values = 1 / (2 * N_values - 1)

# Homozygote and heterozygote frequencies for finite populations
P_homo_finite = p**2 - p * q * f_values
P_hetero_finite = 2 * p * q * (1 + f_values)

# HW expectations
P_homo_HW = p**2
P_hetero_HW = 2 * p * q

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(N_values, P_homo_finite, label='Finite Pop: Homozygote $A_iA_i$', color='red')
plt.plot(N_values, P_hetero_finite, label='Finite Pop: Heterozygote $A_iA_j$', color='blue')
plt.axhline(y=P_homo_HW, color='red', linestyle='--', label='HW Homozygote $p^2$')
plt.axhline(y=P_hetero_HW, color='blue', linestyle='--', label='HW Heterozygote $2pq$')

plt.title('Deviation from Hardy-Weinberg in Finite Populations (p = 0.05)')
plt.xlabel('Population Size (N)')
plt.ylabel('Genotype Frequency')
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```
